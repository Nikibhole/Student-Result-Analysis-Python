# **#Drop Unnamed Column**

```python
fg = fg.drop("Unnamed: 0",axis=1)
print(fg.head())
```

# **#Gender Distribution**

```python
plt.figure(figsize=(5,5))
var=sns.countplot(data=fg,x="Gender", palette="cool")
plt.title("Gender Distribution")
var.bar_label(var.containers[0])
plt.show()
```

#From the above chart we have analysed :
#The number of females in the data is more than the number of males



```python
gb = fg.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Education and Student Scores")
plt.xlabel("Student Scores")
plt.show()
```

#from the above chart we have concluded that the education of parents have good impact on their child scores

```python
gb1 = fg.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parents Marital Status and Student Scores")
plt.xlabel("Student Scores")
plt.show()
```

#from the above chart we have concluded that there is no impact on student score due to their parents marital status

```python
print(fg["EthnicGroup"].unique())
```

```python
groupA = fg.loc[(fg['EthnicGroup'] == 'group A')].count()
groupB = fg.loc[(fg['EthnicGroup'] == 'group B')].count()
groupC = fg.loc[(fg['EthnicGroup'] == 'group C')].count()
groupD = fg.loc[(fg['EthnicGroup'] == 'group D')].count()
groupE = fg.loc[(fg['EthnicGroup'] == 'group E')].count()

l=['group A','group B','group C','group D','group E']
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
c = ['gold','mediumorchid','limegreen', 'tomato', 'deepskyblue']
plt.pie(mlist, labels = l, autopct="%1.2f%%",radius=1.1,colors=c)
plt.title("Distribution Of Ethnic Group")
plt.legend()
plt.show()
```
